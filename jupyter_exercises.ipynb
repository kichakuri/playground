```python
def get_name(whom = "Benard Kichakuri"):
    #Make sure we take care of extra
    whom=whom.strip()
    
    #Capitalize each word
    whom = whom.title()
    
    #Print out the name of the user.
    print(f"Hello {whom}")
    
    return

#Get the name from the user
name = input("What is your name? ")
    
get_name(name)
```
